```python
Image=cv2.imread('/content/drive/MyDrive/dogs/images (1).jpg')
start_point = (100, 50)
end_point = (125, 80)
color = (0, 0, 0)
thickness = -1
image = cv2.rectangle(Image, start_point, end_point, color, thickness)
cv2_imshow(image)
```

```python
# font to put text in ur image
font = cv2.FONT_HERSHEY_SIMPLEX

org = (50, 50)

# fontScale
fontScale = 1

# Blue color in BGR
color = (255, 0, 0)

# Line thickness of 2 px
thickness = 2

# Using cv2.putText() method
image = cv2.putText(Image, '', org, font,
                   fontScale, color, thickness, cv2.LINE_AA)

cv2_imshow(image)
```

```python
center_coordinates = (120, 100)

axesLength = (100, 50)

angle = 0

startAngle = 0

endAngle = 360

color = (0, 0, 255)

thickness = 5

# Using cv2.ellipse() method
# Draw a ellipse with red line borders of thickness of 5 px
image = cv2.ellipse(Image, center_coordinates, axesLength,
           angle, startAngle, endAngle, color, thickness)

cv2_imshow(Image)
```

```python
# ksize
ksize = (20, 30)
image = cv2.blur(Image, ksize, cv2.BORDER_DEFAULT)
cv2_imshow(image)
```

## Task 5 - Resize the image

```python
#to resize the image,to increase the image
img_array = cv2.resize(Image, dsize=(299, 299), interpolation=cv2.INTER_CUBIC)
cv2_imshow(img_array)
```

```python
img_array = cv2.resize(Image, dsize=(100, 100), interpolation=cv2.INTER_CUBIC)
cv2_imshow(img_array)
```
